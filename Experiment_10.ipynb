```python
import pandas as pd
s = pd.Series([10,20,30,40])
print(s)
```

```python
#Create data frame
data = {
    "NAME" : ["A" , "B" , "C"],
    "MARKS" : [30,20,10]
}
df = pd.DataFrame(data)
print(df , "\n")
print(df.loc[0,"NAME"],"\n")
print(df.iloc[2,1])
```

```python
# Find the structure of the Data Frame
print(df.shape)
print(df.ndim)
print(df.size)
```

```python
#Display column names
print(df.columns , "\n")
#Check data types of each column
print(df.dtypes , "\n")
#Acess Individual column
print(df["MARKS"] , "\n")
```

```python
# Access name of first row
print(df.loc[0] , "\n")
# Access name of second row
print(df.loc[1] , "\n")
```

```python
#ADD  a new column
df["GRADE"] = ["FIRST CLASS" , "DISTINCTION" , "SECOND CLASS"]
print(df , "\n")
#Change the marks of A to 40
df.loc[0,"MARKS"] = 40
print(df , "\n")
#Delete a column
del df["GRADE"]
print(df , "\n")
```

```python
#Perform Basic Stastical Analysis
mean1 = df["MARKS"].mean()
print(mean1 , "\n")
max1 = df["MARKS"].max()
print(max1 , "\n")
min1 = df["MARKS"].min()
print(min1)
```

```python
#Apply condition (Filtering)
# Display students whose marks > 10
print(df[df["MARKS"]>10])
```
